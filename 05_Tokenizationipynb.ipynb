# 🧠 Tokenization in Transformers

Tokenization is the process of converting text into smaller chunks (tokens) that models can understand. It’s a fundamental step before feeding any input to a language model.

Transformers don’t understand raw text — they understand numbers. Tokenization bridges that gap.


## 🔹 Why is Tokenization Important?

Language models operate on numbers, not strings. So we need:
- A consistent way to break text into units (tokens).
- A mapping from each token to a numerical ID.
- A method that can handle out-of-vocabulary words.

Most modern LLMs use **subword tokenization** — a balance between word and character-level.


## 🔹 Types of Tokenization

| Type             | Example Input       | Output Tokens              |
|------------------|---------------------|----------------------------|
| Word-level       | "Machine learning"  | ["Machine", "learning"]    |
| Character-level  | "AI"                | ["A", "I"]                 |
| Subword-level    | "unhappiness"       | ["un", "happi", "ness"]    |

Subword-level is most used in LLMs (e.g., BPE, WordPiece).


```python
# ✅ Code: Using HuggingFace tokenizer
from transformers import AutoTokenizer

# Load tokenizer (GPT-2 uses BPE)
tokenizer = AutoTokenizer.from_pretrained("gpt2")

text = "Transformers are powerful."

# Tokenize into subword units
tokens = tokenizer.tokenize(text)
token_ids = tokenizer.convert_tokens_to_ids(tokens)

print("Tokens:", tokens)
print("Token IDs:", token_ids)
```

## 🔍 What's Happening Behind the Scenes?

1. **Tokenize**: `"Transformers are powerful."` → ['Transform', 'ers', ' are', ' powerful', '.']
2. **Map to IDs**: Every token is mapped to a unique integer ID from the vocabulary.
3. These IDs are what go into the model.

Note: GPT-2 uses Byte-Pair Encoding (BPE) for efficient subword splitting.



```python
# ✅ Optional Visualization: Tokens vs IDs
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 4))
plt.bar(tokens, token_ids, color='skyblue')
plt.title("Token vs Token ID")
plt.xlabel("Tokens")
plt.ylabel("Token IDs")
plt.xticks(rotation=45)
plt.grid(True, axis='y')
plt.show()
```

##  Summary

- Tokenization is the first step in the LLM pipeline.
- Subword tokenization allows models to handle unknown or rare words.
- HuggingFace makes tokenization seamless for most pretrained models.
